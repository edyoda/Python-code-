```python
price = soup.find("div",attrs={"class":"m-srp-card__price"})
print(price.text)
```

```python
prices = soup.find_all("div",attrs={"class":"m-srp-card__price"})
p = [price.text for price in prices]
print(p)
```

```python
card = soup.find("div",attrs={"class":"m-srp-card__container"})
price  = card.find("div",attrs={"class":"m-srp-card__price"})
title = card.find("a",attrs={"class":"m-srp-card__title"})
title_text = title.text.replace("\n"," ")
super_area = card.find("div",attrs={"class":"m-srp-card__summary__info"})
data = "{} {} {}".format(title_text,price.text,super_area.text)
print(data)
```

```python
cards = soup.find_all("div",attrs={"class":"m-srp-card__container"})

for card in cards:
    price  = card.find("div",attrs={"class":"m-srp-card__price"})
    title = card.find("a",attrs={"class":"m-srp-card__title"})
    link = title.get("href")
    title_text = title.text.replace("\n"," ")
    super_area = card.find("div",attrs={"class":"m-srp-card__summary__info"})
    
    response2 = requests.get(link)
    soup2 = BeautifulSoup(response2.content,"html.parser")
    owner = soup2.find("p",attrs={"class":"CA_agent_name"})
    owner_text = owner.text.replace("\n","")
    
    
    data = "{} {} {} {}".format(title_text,price.text,super_area.text,owner_text)
    print(data)
```

```python
response = requests.post('https://www.magicbricks.com/property-for-sale/residential-real-estate?proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment&cityName=Bangalore',data=payload)
print(response)
```
